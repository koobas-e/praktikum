Создаём сводную таблицу из DataFrame df, агрегируя данные по "№ товара" и "Наименование товара" с разбивкой по "Плановая дата поставки". Для "Заказано" и "Поставлено" суммируются значения, а для "SL, %" вычисляется среднее. Отсутствующие значения заменяются на 0, добавляются итоговые строки и столбцы.

```python
df_svod = pd.pivot_table(
    df, 
    index=["№ товара", "Наименование товара"],
    columns=["Плановая дата поставки"],
    values=["Заказано", "Поставлено", "SL, %"],
    aggfunc={
        "Заказано": np.sum,
        "Поставлено": np.sum,
        "SL, %": np.mean
    }, 
    fill_value=0, 
    margins=True,
    margins_name='Итог'
)

# Упорядочиваем столбцы так, чтобы каждая плановая дата поставки содержала данные "Заказано", "Поставлено", "SL, %"
df_svod = df_svod.reorder_levels([1, 0], axis=1).sort_index(axis=1, level=0)
```
